```python
# df= SOLUTION DATAFRAME
# n: Total No of Persons affected due to flight Cancellation
# s: {'1_1','1_M','M_1','M_M'}

def solution_ranking(df,n,s):
    score=0
    no_without_alt=0
    mean_arrival_delay=0.0

    dic_sol={'1_1':250,'1_M':150,'M_1':100}

    dict={'BC':12000, 'FC':1000,'PC':500,'EC':250}

    scr_class=0

    for i in range(len(df)):
        mean_arrival_delay=mean_arrival_delay + df.loc[0,'Arrival_TimeDifference']
        scr_class=dict[df.loc[i,'New_ClassName']]-(dict[df.loc[i,'Prev_ClassName']])

    scr_class=scr_class/n
    # print(mean_arrival_delay)
    if(len(df)!=0):
        mean_arrival_delay=mean_arrival_delay/(len(df))
    # print(mean_arrival_delay)
    score+=float((n-len(df)))/float(n)*(100)-mean_arrival_delay*100+100*scr_class+(len(df))*dic_sol[s]/n
    return score
```

```python
# n: total no of affected persons
n=32
sol=['sol_1','sol_2','sol_3','sol_4']
score=[0,0,0,0]
score[0]=solution_ranking(sol_1,n,'1_1')
score[1]=solution_ranking(sol_2,n,'M_1')
score[2]=solution_ranking(sol_3,n,'1_1')
score[3]=solution_ranking(sol_4,n,'M_1')

solution_rank=pd.DataFrame(columns=['Name','Score'])
solution_rank['Name']=sol
solution_rank['Score']=score
solution_rank.sort_values(by='Score',ascending=False,inplace=True)
```
